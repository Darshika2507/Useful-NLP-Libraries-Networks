**Q.2 What is TextBlob and how does it simplify common NLP tasks like sentiment analysis and translation?**

      --> TextBlob is an open-source Python library designed to simplify Natural Language Processing (NLP) tasks by providing an intuitive, user-friendly interface built on top of NLTK and Pattern. It enables rapid prototyping for tasks like sentiment analysis (polarity/subjectivity), translation, tokenization, spelling correction, and part-of-speech tagging with minimal code.

      How TextBlob Simplifies NLP Tasks

      Sentiment Analysis: TextBlob provides immediate, pre-trained sentiment analysis. By calling .sentiment on a text object, it returns a named tuple containing polarity (ranging from -1.0 for negative to 1.0 for positive) and subjectivity (0.0 for objective to 1.0 for subjective).

     Example: TextBlob("I love this!").sentiment-->Sentiment(polarity=0.625, subjectivity=0.6).

     Translation and Language Detection: Using the Google Translate API, TextBlob allows for easy language detection and translation with single method calls like detect_language() and translate(to='es').

     Text Processing Pipeline: It streamlines preprocessing, such as converting text into words/sentences (tokenization), noun phrase extraction, and spelling correction (.correct()).

     Ease of Use: It treats text as Python strings that have learned NLP, allowing for straightforward manipulations.

     Key Features:

     Installation: Installed via pip install textblob and python -m textblob.download_corpora.

     Performance: It is generally faster than NLTK for basic tasks and ideal for applications with limited computational resources.

     Flexibility: While it includes pre-trained models, it also supports custom classifiers (e.g., Naive Bayes).

**Q.4 Describe the architecture and functioning of a Recurrent Natural Network (RNN).**

      --> Recurrent Neural Networks (RNNs) are a class of artificial neural networks designed to process sequential data (e.g., text, audio, time-series) by using internal "memory" to retain information from previous inputs. Unlike feedforward networks, RNNs include feedback loops within their hidden layers, allowing information to persist and influence future outputs, enabling them to handle inputs of variable length.

      Architecture of an RNN

      The architecture of a standard RNN is built upon recurrent neurons, which can be visualized as a chain-like structure.

      Hidden State (): The core memory component that summarizes the sequence information seen so far.

      Feedback Loops (Recurrence): At each time step (), the hidden layer receives two inputs: the current input () and the hidden state from the previous time step ().

      Parameter Sharing: The same weights () are shared across all time steps, reducing model complexity and allowing it to recognize patterns regardless of their position in the sequence.

      Unfolded Structure: When visualized over time, the network "unfolds" into a chain, where each step corresponds to a specific point in time ().

      Functioning and Operation

      1. Input: The network receives input for the current time step.

      2. Hidden State Update: The new hidden state h1 is computed using the formula:h1 = tanh(Wxhx1 + Whhh-1 + bh). This uses the current input and the previous hidden state (memory).

      3. Output Generation: The output y1 is generated using the new hidden state, often via a softmax function: y1 = softmax(Whyh1 + by).

      4. Looping: The new hidden state h1 is passed as input to the next time step, enabling temporal memory.

      Training: Backpropagation Through Time (BPTT)

      RNNs are trained using a method called Backpropagation Through Time (BPTT).

      Forward Pass: Data is fed through the sequence, and hidden states are calculated for all time steps.

      Loss Calculation: The error is calculated at each step (or at the end of the sequence).

      Backward Pass: The errors are propagated backward through time (from step T back to 1) to update the weights based on all preceding steps.

      Challenges and Variations

      Standard RNNs struggle with long-term dependencies due to the vanishing and exploding gradient problems, where signals from long ago disappear, causing the network to "forget".

      LSTM (Long Short-Term Memory): Addresses this with a cell state and three gates (input, forget, output) to manage memory effectively.

      GRU (Gated Recurrent Unit): A simpler alternative to LSTM that uses update and reset gates, allowing for faster training.

      Bidirectional RNNs (BRNN): Process data in both forward and backward directions, allowing the network to use both past and future context for predictions.

      Key Applications

      Natural Language Processing (NLP): Sentiment analysis, text generation, machine translation.

     Time-Series Forecasting: Stock price prediction, weather forecasting.

     Speech Recognition: Converting audio into text.

**Q.5 What is the key difference between LSTM and GRU networks in NLP applications?**

      --> The key difference between LSTM and GRU in NLP is that LSTM uses three gates (input, forget, output) and a separate cell state for complex, long-term memory, while GRU uses two gates (reset, update) and merges the cell/hidden states for faster, more efficient training.

      Key Differences and Applications:

      Architecture Complexity: LSTM has a more complex structure, making it better for capturing long-term dependencies in long, complex sequences.

      Computational Efficiency: GRU is simpler, has fewer parameters, and trains faster, making it better suited for smaller datasets or resource-constrained environments.

      Performance: LSTMs are generally preferred for complex, long-range dependencies (e.g., machine translation), whereas GRUs provide similar performance on shorter sequences with faster, more efficient training (e.g., sentiment analysis).

**Q.6 Write a Python program using TextBlob to perform sentiment analysis on the following paragraph of text:**

**“I had a great experience using the new mobile banking app. The interface is intuitive, and customer support was quick to resolve my issue. However, the app did crash once during a transaction, which was frustrating"**

**Your program should print out the polarity and subjectivity scores.**


```python
# Import TextBlob
from textblob import TextBlob

# Paragraph text
text = """I had a great experience using the new mobile banking app.
The interface is intuitive, and customer support was quick to resolve my issue.
However, the app did crash once during a transaction, which was frustrating."""

# Create TextBlob object
analysis = TextBlob(text)

# Get sentiment
polarity = analysis.sentiment.polarity
subjectivity = analysis.sentiment.subjectivity

# Print results
print("Polarity Score:", polarity)
print("Subjectivity Score:", subjectivity)
```

**Q.7  Given the sample paragraph below, perform string tokenization and frequency distribution using Python and NLTK:**

**“Natural Language Processing (NLP) is a fascinating field that combines linguistics,computer science, and artificial intelligence. It enables machines to understand, interpret, and generate human language. Applications of NLP include chatbots, sentiment analysis, and machine translation. As technology advances, the role of NLP in modern solutions is becoming increasingly critical.”**

```python
# Import required libraries
import nltk
from nltk.tokenize import word_tokenize
from nltk.probability import FreqDist

# Download tokenizer (run once)
nltk.download('punkt')
nltk.download('punkt_tab') # Added to resolve the LookupError

# Sample paragraph
text = """Natural Language Processing (NLP) is a fascinating field that combines linguistics,
computer science, and artificial intelligence. It enables machines to understand,
interpret, and generate human language. Applications of NLP include chatbots,
sentiment analysis, and machine translation. As technology advances, the role of NLP
in modern solutions is becoming increasingly critical."""

# Tokenization (split text into words)
tokens = word_tokenize(text)

# Convert to lowercase and remove punctuation
words = [word.lower() for word in tokens if word.isalpha()]

# Frequency Distribution
freq_dist = FreqDist(words)

# Print tokens
print("Tokens:")
print(words)

# Print frequency distribution
print("\nFrequency Distribution:")
for word, freq in freq_dist.items():
    print(word, ":", freq)
```

**Q.8 Implement a basic LSTM model in Keras for a text classification task using the following dummy dataset. Your model should classify sentences as either positive (1) or negative (0).**

**# Dataset texts = [ “I love this project”, #Positive**
**“This is an amazing experience”, #Positive**
**“I hate waiting in line”, #Negative**
**“This is the worst service”, #Negative**
**“Absolutely fantastic!” #Positive**
]
**

**labels = [1, 1, 0, 0, 1]**

**Preprocess the text, tokenize it, pad sequences, and build an LSTM model to train on this data. You may use Keras with TensorFlow backend.**

```python
import numpy as np
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, LSTM, Dense

texts = [
"I love this project",            # Positive
"This is an amazing experience",  # Positive
"I hate waiting in line",         # Negative
"This is the worst service",      # Negative
"Absolutely fantastic!"           # Positive
]

labels = [1, 1, 0, 0, 1]

tokenizer = Tokenizer()
tokenizer.fit_on_texts(texts)

sequences = tokenizer.texts_to_sequences(texts)

print("Sequences:", sequences)

max_length = max(len(seq) for seq in sequences)

padded_sequences = pad_sequences(sequences, maxlen=max_length, padding='post')

print("Padded Sequences:\n", padded_sequences)

vocab_size = len(tokenizer.word_index) + 1

model = Sequential()

model.add(Embedding(input_dim=vocab_size, output_dim=16, input_length=max_length))

model.add(LSTM(16))

model.add(Dense(1, activation='sigmoid'))

model.fit(
    padded_sequences,
    np.array(labels),
    epochs=20,
    verbose=1
)

test_sentence = ["I love this service"]

test_seq = tokenizer.texts_to_sequences(test_sentence)
test_pad = pad_sequences(test_seq, maxlen=max_length, padding='post')

prediction = model.predict(test_pad)

print("Prediction:", prediction)

model.summary()
```
